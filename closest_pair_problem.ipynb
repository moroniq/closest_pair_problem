```python
#Closest pair problem - The closest pair of points problem or closest pair problem is a problem of computational geometry: given n points in metric space, 
#find a pair of points with the smallest distance between them.
```

```python
points = [(0,0), (0,1), (3,4), (2,8)]
```

```python
def closest_pair(points):
    best_dist2 = float("inf") #nieskończoność
    best_pair = None
    n = len(points)

    for i in range(n):
        for j in range(i+1, n):
            dx = points[j][0] - points[i][0]
            dy = points[j][1] - points[i][1]
            dist2 = dx*dx + dy*dy
    
            if dist2 < best_dist2:
                best_dist2 = dist2
                best_pair = (points[i], points[j])

    print(f"Best pair is: {best_pair}")
    print(f"Best distance sqauare is: {best_dist2}")    
```

```python
closest_pair(points)
```
